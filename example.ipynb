# Article similarity

Program can detect relations between articles.

### The algorith is following:
- find in google search engine key word,
- download finded article,
- compare content of them,
- make charts of computed data.

### Used 

### How use function make_analize ?
    Function need get key words as first argument,
    second argument is optional - define how many results in google will be taken (rounded to 10).

### How interpret results?
    If searched information is popular then finded articles will contain similar content - about 10% - 80%.
    Else content will be not related and there will similatity of 0% - 10%.
    To avoid completly duplicated articles, 100% identical aricle has 0% similarity.
    Program should have greatest efficiency on sities with english language.
